● Step 1: Write a function that takes a list of numbers and returns the mean, min, and max as a dictionary.


```python
def analayz_numbers(numbers):
    mean = sum(numbers) / len(numbers)
    return{
     "mean": mean,
     "min": min(numbers),
     "max": max(numbers)
    }
```

# Testing the function and pass a list of numbers

```python
numbers=[10,20,30,40,50]
result = analayz_numbers(numbers)
print(result)
```

## Filtering Even Numbers Using List Comprehension

This cell uses a single list comprehension to filter even numbers from a given list.
The condition checks whether each number is divisible by 2 and keeps only even values.

```python
numbers=[1,2,3,4,5]
even_numbers=[value for value in numbers     if value % 2 == 0 ]
print(even_numbers)
```

## Define Student Class

This cell creates a Student class with attributes for name, age, and GPA.
The class also contains a method to display student information.

```python
class student:
    def __init__(self,name,age,gpa):
        self.name=name
        self.age=age
        self.gpa=gpa
    def info(self):
        print(f"my name is {self.name}")
        print(f"im {self.age} years old")
        print(f"my gpa {self.gpa}")

```
