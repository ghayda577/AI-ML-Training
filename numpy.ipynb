## NumPy Array Manipulation

● Step 1: Create a 2D array of shape (4, 4) filled with values 1-16 and print its shape and dtype.


```python
import numpy as np 

arr2d = np.arange(1,17).reshape(4,4)
print (arr2d)
print("shape:" , arr2d.shape)
print("Data type:" , arr2d.dtype)
```

● Step 2: Use slicing to extract the second column and the last row.


```python
second_col= arr2d[: ,1]
last_row = arr2d[-1]
print("second columon =" , second_col)
print("last row =" , last_row)
```

● Step 3: Use a boolean mask to select all values greater than the array's mean.


```python
mean = arr2d.mean()
print(mean)
print("values grater than mean :", arr2d[arr2d > mean])
```

● Step 4: Add a 1D row array to every row of a 2D array using broadcasting, and verify the result manually.


```python
arr = np.array([1 ,2 ,3 ,4])
result =arr2d +arr
print("original array :\n" ,arr2d)
print("row array :\n", arr)
print("added arrays :\n", result)
```
